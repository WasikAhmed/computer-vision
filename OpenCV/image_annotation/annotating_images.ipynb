# Image Annotation

## Importing Libraries

````python
import os
import cv2
import matplotlib
import numpy as np
import matplotlib.pyplot as plt

from zipfile import ZipFile
from urllib.request import urlretrieve

%matplotlib inline
````

## Downloading Assets

````python
def download_and_unzip(url, zip_file, folder):
    print(f"Downloading and extracting assests....", end="")
    
    # Create the folder if it does not exist
    if not os.path.exists(folder):
        os.makedirs(folder)
    
    try:
        # Download the file if it does not exist
        if not os.path.exists(zip_file):
            urlretrieve(url, zip_file)
        
        # Unzip the file
        with ZipFile(zip_file, 'r') as zip_ref:
            zip_ref.extractall(folder)

        print('Done')
    
    except Exception as e:
        print(f"Error: {e}")
````

````python
URL = r"https://www.dropbox.com/s/48hboi1m4crv1tl/opencv_bootcamp_assets_NB3.zip?dl=1"
folder = os.path.join(os.getcwd(), 'assets')
zip_file = os.path.join(folder, 'opencv_bootcamp_assets_NB3.zip')

download_and_unzip(URL, zip_file, folder)
````

## Read the Image

````python
image_bgr = cv2.imread('assets/Apollo_11_Launch.jpg', cv2.IMREAD_COLOR)
image_rgb = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2RGB)

# Display the image
plt.imshow(image_rgb)
````

## Drawing a Line

Line is drawn using the `cv2.line()` function.

Syntax:
```
img = cv.line(img, pt1, pt2, color[, thickness[, lineType[, shift]]])
```

````python
image_line = image_bgr.copy()
line_color = (255, 0, 0) # Blue color in BGR

cv2.line(image_line, pt1=(200, 100), pt2=(400, 100), color=line_color, thickness=5, lineType=cv2.LINE_AA)

# Display the image
plt.imshow(cv2.cvtColor(image_line, cv2.COLOR_BGR2RGB))
````

## Drawing a Circle

Circle is drawn using the `cv2.circle()` function.

Syntax:
```
img = cv.circle(img, center, radius, color[, thickness[, lineType[, shift]]])
```

````python
img_circle = image_bgr.copy()
circle_center = (900, 500)
circle_radius = 100
circle_color = (0, 255, 0) # Green color in BGR

cv2.circle(img_circle, center=circle_center, radius=circle_radius, color=circle_color, thickness=5, lineType=cv2.LINE_AA)

# Display the image
plt.imshow(cv2.cvtColor(img_circle, cv2.COLOR_BGR2RGB))
````

## Drawing a Rectangle

Rectangle is drawn using the `cv2.rectangle()` function.

Syntax:
```
img = cv.rectangle(img, pt1, pt2, color[, thickness[, lineType[, shift]]])
```

````python
image_rectangle = image_bgr.copy()
pt1 = (500, 100)
pt2 = (700, 600)
rectangle_color = (0, 0, 255) # Red color in BGR

cv2.rectangle(image_rectangle, pt1=pt1, pt2=pt2, color=rectangle_color, thickness=5, lineType=cv2.LINE_AA)

# Display the image
plt.imshow(cv2.cvtColor(image_rectangle, cv2.COLOR_BGR2RGB))
````

## Adding Text

Text is added using the `cv2.putText()` function.

Syntax:
```
img = cv.putText(img, text, org, fontFace, fontScale, color[, thickness[, lineType[, bottomLeftOrigin]]])
```

````python
image_text = image_bgr.copy()
text = "Apollo 11 Saturn V Launch, July 16, 1969"
font_org = (200, 680)
font_scale = 2.5
font_face = cv2.FONT_HERSHEY_PLAIN
font_color = (0, 255, 0) # Red color in BGR
font_thickness = 2

cv2.putText(image_text, text, org=font_org, fontFace=font_face, fontScale=font_scale, color=font_color, thickness=font_thickness, lineType=cv2.LINE_AA)

# display the image
# plt.imshow(cv2.cvtColor(image_text, cv2.COLOR_BGR2RGB))
plt.imshow(image_text[:, :, ::-1])
````
